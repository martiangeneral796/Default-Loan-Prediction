```python
# Customers are potential defaulter or not

# Load Libraries
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

# Create dataset
# Input - Monthly income, Loan amount, credit score, Age, DTI
x = np.array([[50000, 200000, 500, 30, 0.5], [60000, 150000, 650, 33, 1.2], [55000, 600000, 200, 28, 0.7], [45000, 800000, 400, 25, 1.3], [70000, 500000, 800, 29, 1.6]])
# Output - 1 = Default, 0 = Not default
y = np.array([[1], [0], [1], [0], [1]])

# Build model
model = Sequential()
model.add(Dense(8, input_dim = 5, activation ='relu'))
model.add(Dense(1, activation = 'sigmoid'))

# Compile model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Train model
model.fit(x, y, epochs = 300, verbose = 1)

# Make predictions
x_test = np.array([[60000, 150000, 650, 33, 1.2], [45000, 800000, 400, 25, 1.3], [70000, 500000, 800, 29, 1.6]])
prob = model.predict(x_test)
pred = (prob > 0.5).astype(int)

# Print result
for i in range(3):
    print(f'Monthly income :{x_test[i][0]}, Loan amount :{x_test[i][1]}, Credit score :{x_test[i][2]}, Age :{x_test[i][3]}, DTI :{x_test[i][4]}')
    print(f"Probability of Default :{prob[i][0]}")
    print(f"Prediction (1 = Default, 0 = Not default): {pred[i][0]}")
```
