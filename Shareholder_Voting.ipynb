```
# Shareholders will vote in favour of resolutions or not

# Load Libraries
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

# Create data
# Input - DII Stake, FII Stake, Retail Stake
x = np.array([[55, 40, 5], [40, 55, 5], [35, 35, 30]])
# Output - 1 = Favour, 0 = Against
y = np.array([[1], [0], [1]])

# Build Model
model = Sequential()
model.add(Dense(8, input_dim = 3, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

# Compile Model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Train Model
model.fit(x, y, epochs = 300, verbose = 1)

# Make predictions
x_test = np.array([[55, 40, 5], [40, 55, 5], [35, 35, 30]])
prob = model.predict(x_test)
pred = (prob > 0.5).astype(int)

# Print result
for i in range(3):
    print(f'DII Stake :{x_test[i][0]}, FII Stake :{x_test[i][1]}, Retail Stake :{x_test[i][2]}')
    print(f"Probability of Voting :{prob[i][0]}")
    print(f"Prediction (1 = Favour, 0 = Against): {pred[i][0]}")
```
